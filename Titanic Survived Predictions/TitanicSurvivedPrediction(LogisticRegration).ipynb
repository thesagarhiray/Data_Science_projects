```python
import numpy as np
import pandas as pd
```

Step1: Gathering the data

```python
titanic = pd.read_csv("/content/train_and_test2.csv")
titanic.head()
```

```python
titanic.info()
```

Step2: Data Cleaning

```python
titanic_new = titanic
for i in titanic:
  mean=titanic_new[i].mean()
  titanic_new[i].fillna(mean,inplace=True)  # Filling the empty cell with the mean value of that column.
  count = 0
  if(i != '2urvived'):
    for j in titanic[i]:
      count += j
    if count == 0:
      titanic_new = titanic_new.drop(i,axis=1) # Drop all the columns contains only zero values

titanic_new = titanic_new.drop('Passengerid',axis = 1)
titanic_new.head()
```

Step3: Divide the data into independant and dependant variables

```python
x = titanic_new.drop('2urvived',axis=1)
x.head()
```

```python
y = titanic_new['2urvived']
y.head()
```

Step4: Split the data into training and testing

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
x_train.head()
```

```python
x_test.head()
```

```python
y_train.head()
```

```python
y_test.head()
```

#Creating the ML model for Logistic regression

```python
from sklearn.linear_model import LogisticRegression
logRegression = LogisticRegression()
```

```python
logRegression.fit(x_train, y_train)
```

```python
predictions = logRegression.predict(x_test)
print(predictions)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test,predictions)
ac = accuracy_score(y_test,predictions)
```
